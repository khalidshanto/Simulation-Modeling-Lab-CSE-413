```python
import numpy as np
from scipy.stats import kstest, norm
import matplotlib.pyplot as plt
```

```python
# ===== Data Preparation =====
data = np.array([15, 17, 16, 14, 18, 20, 19, 15, 21, 22])
```

```python
# Calculate statistics from data
mu = np.mean(data)
sigma = np.std(data)
```

```python
# Normalize data (z-score transformation)
normalized = (data - mu) / sigma
```

```python
# ===== Statistical Test =====
# Compare against standard normal distribution (mean=0, std=1)
ks_stat, p_value = kstest(normalized, 'norm')
```

```python
# ===== Visualization =====
# Generate points for CDF comparison
x = np.sort(normalized)
y_empirical = np.linspace(0, 1, len(x))  # Simple ECDF
y_normal = norm.cdf(x)
```

```python
plt.figure(figsize=(8, 5))
plt.step(x, y_empirical, where='post', label='Data CDF')
plt.plot(x, y_normal, 'r--', label='Normal CDF')
plt.title("Temperature Distribution Comparison")
plt.xlabel("Normalized Values (Z-scores)")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True)
```

```python
# ===== Results & Conclusion =====
print(f"KS Statistic: {ks_stat:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Normal Parameters: μ = {mu:.1f}, σ = {sigma:.1f}")
```

```python
# Hypothesis test conclusion
sig_level = 0.05
if p_value < sig_level:
    print("Conclusion: Not normally distributed (reject H₀)")
else:
    print("Conclusion: Normally distributed (keep H₀)")

plt.show()
```
